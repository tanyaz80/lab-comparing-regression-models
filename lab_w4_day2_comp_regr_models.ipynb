```python
#In this function I am combining SUV, Sports car, Luxury Car and Luxury SUV in 1 category SUV. 
#As they have very small range in all data set
def combine_vehicle_class (df_original):
    df=df_original.copy()
    df['Vehicle Class'] = np.where(df['Vehicle Class'].isin(['SUV','Sports Car','Luxury SUV', 'Luxury Car']) , 'SUV', df['Vehicle Class'])
    return df

#In this function I am combining disabled, retired group and medical leave in 1 group-retired.
def cleanEmployment(x):
    x = x.lower()
    if 'unemployed' in x:
        return "unemployed"
    elif 'employed' in x:
        return "employed"
    elif 'disabled' in x:
        return 'disabled'
    else:
        return 'disabled'
   
def clean_employment(df_original):
    df=df_original.copy()
    df['EmploymentStatus'] = df['EmploymentStatus'].apply(cleanEmployment)
    return df

#In this function I decided to combine L1,L2 and L3  and not by personal, corporate and special, as special will be still very small to be 
#representative -around 350. 
def combine_policy (df_original):
    df=df_original.copy()
    df['Policy'] = np.where(df['Policy'].isin(['Personal L3','Corporate L3','Special L3']) , 'Combine L3', df['Policy'])
    df['Policy'] = np.where(df['Policy'].isin(['Personal L2','Corporate L2','Special L2']) , 'Combine L2', df['Policy'])
    df['Policy'] = np.where(df['Policy'].isin(['Personal L1','Corporate L1','Special L1']) , 'Combine L1', df['Policy'])
    return df

#In this function all the Masters and Doctors I put into Bachelor part. Good level of education.
def cleanEducation(x):
    x = x.lower()
    if 'bachelor' in x:
        return "bachelor"
    elif 'college' in x:
        return "college"
    elif 'high school or below' in x:
        return 'high school'
    else:
        return 'bachelor'

def clean_education(df_original):
    df=df_original.copy()
    df['Education'] = df['Education'].apply(cleanEducation)
    return df

#special group had a very small value, I added them to corporate group
def cleanPolicyType(x):
    x = x.lower()
    if 'personal auto' in x:
        return "personal auto"
    elif 'corporate auto' in x:
        return "corporate auto"
    else:
        return 'corporate auto'

def clean_policy_type(df_original):
    df=df_original.copy()
    df['Policy Type'] = df['Policy Type'].apply(cleanPolicyType)
    return df


def apply_all(df_original):
    result=clean_employment(df_original)
    result=combine_policy(result)
    result=clean_education(result)
    result=clean_policy_type(result)
    return result
```

```python
categorical_clean_train=apply_all(categoricals_train)
categorical_clean_test=apply_all(categoricals_test)
```

## Great! Now define a function that takes a list of models and train (and tests) them so we can try a lot of them without repeating code.

```python
from sklearn.neighbors import KNeighborsRegressor

def train_test_models(list_of_models, X_train, y_train, X_test,  y_test):                 
                      
    train_R2 = []
    test_R2 = []
 
    for model in list_of_models:
        model.fit(X_train, y_train)
        y_train_pred = model.predict(X_train)
        train_R2.append(r2_score(y_train, y_train_pred))

        y_test_pred = model.predict(X_test)
        test_R2.append(r2_score(y_test, y_test_pred))
        
    return train_R2, test_R2
            
```

## Use the function to check LinearRegressor and KNeighborsRegressor.

```python
list_of_models = [LinearRegression(), KNeighborsRegressor(n_neighbors=3)]

train_R2, test_R2= train_test_models(list_of_models, X_final_train, y_train, X_final_test, y_test)

for i in range(len(list_of_models)):
    print(f'Model: {list_of_models[i]}')
    print(f'    Train-R2: {train_R2[i]}')
    print(f'    Test-R2:  {test_R2[i]}')
```

## You can check also the MLPRegressor for this task!

```python
from sklearn.neural_network import MLPRegressor

list_of_models = [LinearRegression(), KNeighborsRegressor(n_neighbors=3), MLPRegressor() ]

train_R2, test_R2= train_test_models(list_of_models, X_final_train, y_train, X_final_test, y_test)

for i in range(len(list_of_models)):
    print(f'Model: {list_of_models[i]}')
    print(f'    Train-R2: {train_R2[i]}')
    print(f'    Test-R2:  {test_R2[i]}')

```
